```python
import folium
import pandas as pd
```

```python
shapes = pd.read_csv("../../data/raw/20250518/shapes.txt")
```

```python
shapes.head(100)
```

```python
# egy shape_id kiválasztása (pl. az első)
shape_id = shapes['shape_id'].iloc[0]
shape_points = shapes[shapes['shape_id'] == shape_id].sort_values(by='shape_pt_sequence')

# térkép középpontja (első pont)
center = [shape_points['shape_pt_lat'].iloc[0], shape_points['shape_pt_lon'].iloc[0]]
m = folium.Map(location=center, zoom_start=12)

# vonal kirajzolása
coords = list(zip(shape_points['shape_pt_lat'], shape_points['shape_pt_lon']))
folium.PolyLine(coords, color="blue", weight=4.5, opacity=0.8).add_to(m)

m  # notebookban megjelenik
```
